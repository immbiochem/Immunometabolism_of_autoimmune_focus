### Psoriasis

#### Main source:

https://www.ebi.ac.uk/ena/browser/view/PRJNA635048?show=reads

**Project: PRJNA635048**

Single-cell RNA sequencing is transforming how we understand skin immunology, but previous human skin single-cell RNA sequencing data included only a small fraction of inflammatory cells among the overall cell population, such that functional subsets may be difficult to ascertain. We have overcome these obstacles by harvesting inflammatory cells emigrating from a half of 6 mm punch biopsy skin after 48-hour incubation in culture medium without any enzyme, and then analyzing the harvested cells with single-cell RNA sequencing. By this strategy, we obtained single-cell RNA sequencing data of 24,354 cells (leukocytes 46.0%, keratinocytes 49.6%, and melanocytes 2.4%) from 13 human psoriasis skin and 5 healthy volunteer skin. Unsupervised clustering identified NK cells, CD161+ T-cells, CD8+ T-cells, CD4+ T-cells, regulatory T-cells, mature & semimature dendritic cells, melanocytes, and keratinocytes in different layers - S. (Stratum) corneum, S. granulosum, S. spinosum, and S. basale. To understand psoriasis immunopathogenesis at single-cell levels, we compared gene expression between psoriasis cells vs. control cells within each inflammatory cell subtype clusters. Overall design: Transcriptome profiles of 13 human psoriasis lesional skin and 5 healthy volunteers' normal control skin were generated by droplet-based single cell RNA-sequencing.

#### Additional source:

https://www.ebi.ac.uk/ena/browser/view/PRJNA816908?show=reads

**Project: PRJNA816908**

To understand immunological differences between recurred and non-recurred psoriatic lesion after cease of treatment with ixekizumab Overall design: We obtained 5-mm biopsied skin samples from recurred site and non-recurred site. Skin tissue was processed for single cell RNA sequencing.

### Rheumatoid arthritis

#### Main source:

https://www.ebi.ac.uk/ena/browser/view/PRJNA803484?show=reads

**Project: PRJNA803484**

Ectopic lymphoid structures (ELS) can develop in rheumatoid arthritis (RA) synovial tissue, but the precise pathways of B cell activation and selection are not well understood. Here, we identified a unique B cell population in the synovium characterized by co-expression of a family of orphan nuclear receptors, NR4A1 (also known as NUR77), NR4A2 (NURR1) and NR4A3 (NOR1), that is highly enriched at both early and late stages of RA. See doi:10.1101/2021.05.14.443150 for details. Overall design: B cells from Synovium (4 donors) were disaggregated by combination of an enzymatic digestion and mechanical disruption. For one matched donor, B cells were sorted from peripheral blood. B cells were gated as CD45+CD3-CD14-CD19+.


#### PBMC:

https://www.ebi.ac.uk/ena/browser/view/PRJNA667720?show=reads

**Project: PRJNA667720**

The transcriptome of PBMC from rheumatoid arthritis patient hasn't been compenhensively profiled, and heterogeneous characteristics of blood monocytes in RA patients are much unknown. We performed the single cell transcriptomic analysis of PBMC from rheumatoid arthritis (RA) patient, and monocyte populations were extracted during the analysis. CD127 expression associated expression pattern of inflammatory genes was identified in RA patient's blood monocytes. Overall design: PBMC were isolated from peripheral blood of RA patient by Ficoll gradients, and isolated PBMC were frozen for preservation and sent out for scRNA-seq library construction by using 10x genomics platform.

### SLE

#### Main source:

https://www.ebi.ac.uk/ena/browser/view/PRJNA682238?show=reads

**Project: PRJNA682238**

To investigate ISGs expression chracteristics of leucocyte types of SLE patients, single cell RNA sequencing was applied to 2 SLE patients and 1 health volunteer. Overall design: 10X Genomics single cell RNA sequencing was applied to PBMC of 2 SLE new onset patients without using any immunosupressive drugs and 1 health volunteer.
